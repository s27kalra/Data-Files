```python
# Renaming column names
new_col_dict = {'timestamp' : 'time',
                'cnt': 'count',
                't1' : 'Real_temp_c',
                't2' : 'feels_like_temp_c',
                'hum' : 'Humidity_Percent',
                'wind_speed' : 'Wind_Speed',
                'weather_code' : 'weather',
                'is_holiday' : 'Holiday',
                'is_weekend' : 'Weekend',
                'season' : 'Season'
               }
Bikes.rename(new_col_dict, axis=1, inplace=True)
```

```python
# Creating Season Dictionary
Season_dict = { '0.0': 'spring',
               '1.0': 'summer',
               '2.0' : 'fall',
               '3.0': 'winter'
              }
Bikes.rename(Season_dict, axis=1, inplace=True)
```

```python
# Creating Weather Dictionary
Weather_dict = {
    '1.0' : 'Clear',
    '2.0' : 'Scattered clouds',
    '3.0' : 'Broken clouds',
    '4.0' : 'Cloudy',
    '7.0' : 'Rain',
    '10.0' : 'Rain with Thuderstrom',
    '26.0' : 'Snowfall'
}

Bikes.rename(Weather_dict, axis=1, inplace=True)
```

```python
# Changing the humidity values to percentage
Bikes.Humidity_Percent = Bikes.Humidity_Percent/100
```

```python
# changing the seasons column data type to string
Bikes.Season = Bikes.Season.astype('str')
# Mapping the values to actual season name
Bikes.Season = Bikes.Season.map(Season_dict)
```

```python
# Changing the weather column data type to string
Bikes.weather = Bikes.weather.astype('str')
# Mapping weather values to actual written values
Bikes.weather = Bikes.weather.map(Weather_dict)
```
